```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
a=pd.read_csv('titanic.csv')
a.head()
b=a[['Age','Sex']].dropna(subset=['Age'])
plt.figure(figsize=(10,6))
sns.histplot(data=b,x='Age',hue='Sex',multiple='stack',bins=30, palette='Set2')
plt.title('Age Distribution by Sex')
plt.xlabel('Age')
plt.ylabel('Count')     
plt.tight_layout()
plt.show()
```
